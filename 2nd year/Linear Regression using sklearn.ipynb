```python
import numpy as np
import pandas as pd
data=pd.read_csv("headbrain.csv")
data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x = data['Head Size(cm^3)']
x=np.array(x)
x=x.reshape(-1,1)
y=data['Brain Weight(grams)']
y=np.array(y)
x,y
```

```python
reg=LinearRegression()
reg.fit(x,y)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
y_pred=reg.predict(x)
```

```python
y_pred
```

```python
y
```

```python
rmse=np.sqrt(mean_squared_error(y,y_pred))
rmse
```

```python
r2=reg.score(x,y)
r2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y) #plot point of  original data
```

```python
X=x
y_pred2=reg.intercept_+reg.coef_*X   #y=mx+c
```

```python
plt.plot(x,y_pred2,color="orange")
```
