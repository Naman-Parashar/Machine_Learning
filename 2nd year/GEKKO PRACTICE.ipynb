```python
import numpy as np
from gekko import GEKKO
tm=np.linspace(0,20,100)
m=GEKKO(remote=False)  # remote for cloud solve or local solve.
m.time=tm
k=10
y=m.Var(value=5.0)
t=m.Param(value=tm)
m.Equation([k*y*dt() == -t**3*y])
m.options.IMODE=4
m.solve(disp=False)
```
