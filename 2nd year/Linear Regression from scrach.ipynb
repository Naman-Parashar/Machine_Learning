```python
x = data['Head Size(cm^3)']
```

```python
y = data['Brain Weight(grams)']
```

```python
x1 = np.array(x)
y1 = np.array(y)
n1 = x.size
n2 = y.size
x1,y1
```

```python
xsum = sum(x1)
ysum = sum(y1)
xsum,ysum
```

```python
xmean = xsum/n1
ymean = ysum/n2
xmean,ymean
```

```python
s1=0
s2=0
for i in range(len(x1)):
  s1 = s1+(x1[i]-xmean)*(y1[i]-ymean)
for j in range(len(x1)):
  s2 = s2+(pow((x1[j]-xmean),2))
m=s1/s2
m  
  
```

```python
c=ymean-(m*xmean)
c
```

```python
print("Y=",m,"X +",c)
```

```python
for i in range(len(x1)):
    y=(m*x1[i])+c
    print(y)
```
